## Visualizations

### 1. Bar Plot of Total Letter Frequencies Across Dataset

```python
letter_columns = [col for col in df.columns if col != 'Label']
total_counts = df[letter_columns].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=total_counts.index, y=total_counts.values, palette='viridis')
plt.title('Total Frequency of Each Letter (A-Z)')
plt.ylabel('Total Count')
plt.xlabel('Letter')
plt.show()
```

###  2. Boxplot of Each Letter Frequency

```python
plt.figure(figsize=(14,6))
sns.boxplot(data=df[letter_columns])
plt.xticks(rotation=90)
plt.title('Boxplot of Letter Frequencies (A-Z)')
plt.ylabel('Frequency')
plt.xlabel('Letter')
plt.show()
```

### 3. Correlation Heatmap Between Letters

```python
plt.figure(figsize=(14,10))
corr_matrix = df[letter_columns].corr()
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap Between Letter Frequencies')
plt.show()
```

###  4. Countplot of Target Classes

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Label', data=df, palette='Set2')
plt.title('Target Class Distribution')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()
```

### 5. Violin Plot: Frequency of Letter 'A' by Target Class

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='Label', y='A', data=df, palette='muted')
plt.title("Distribution of 'A' by Target Class")
plt.show()
```

### 6. Pairplot of Top 4 Frequent Letters Colored by Target

```python
top_letters = total_counts.head(4).index.tolist()

sns.pairplot(df[top_letters + ['Label']], hue='Label', palette='husl')
plt.suptitle("Pairplot of Top 4 Frequent Letters", y=1.02)
plt.show()
```

### 7. KDE Plot of Letter 'E' Across Target Classes

```python
plt.figure(figsize=(8,5))
for label in df['Label'].unique():
    sns.kdeplot(df[df['Label'] == label]['E'], label=f'Target {label}', fill=True)

plt.title("KDE Plot for Letter 'E' by Target")
plt.xlabel('Frequency of E')
plt.legend()
plt.show()
```

### 8. Heatmap of Mean Letter Frequencies Per Class

```python
mean_by_class = df.groupby('Label')[letter_columns].mean()

plt.figure(figsize=(14,8))
sns.heatmap(mean_by_class, cmap='YlGnBu', annot=True, fmt='.1f')
plt.title('Mean Letter Frequency by Target Class')
plt.ylabel('Target')
plt.xlabel('Letter')
plt.show()
```

### 9. Histogram of Letter 'T'

```python
plt.figure(figsize=(8,5))
sns.histplot(df['T'], bins=30, color='coral', kde=True)
plt.title("Histogram of Letter 'T'")
plt.xlabel('Frequency of T')
plt.ylabel('Count')
plt.show()
```
