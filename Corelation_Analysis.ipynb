### Corelation Analysis

### Calculate Correlation Matrix

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load preprocessed dataset
df = pd.read_csv('genome_sequences.csv')  # update this if you renamed the processed file

# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)
```

###  Visualize the Correlation Matrix

```python
plt.figure(figsize=(12, 10))
# Create heatmap
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Genome Dataset')
plt.show()
```

### Identifying Significant Correlations

```python
threshold = 0.7
# Find pairs with high correlation (excluding self-correlation)]
strong_corrs = corr_matrix[(abs(corr_matrix) > threshold) & (corr_matrix != 1.0)]
# Stack to see in pair format
significant_pairs = strong_corrs.stack().reset_index()
significant_pairs.columns = ['Feature 1', 'Feature 2', 'Correlation']
print("Significant Correlation Pairs (|r| > 0.7):")
print(significant_pairs)
```
